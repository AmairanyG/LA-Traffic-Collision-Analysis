```python
#import modules for the analysis environment
import gmaps
import pandas as pd
from ast import literal_eval
import matplotlib.pyplot as plt

# Google developer API key
from config import gkey

# Configure gmaps
gmaps.configure(api_key=gkey)
```

```python
df = pd.read_csv("Data/clean_data/clean_data.csv", index_col=None)
df.head()
```

```python
drop_df = df.drop("Unnamed: 0", )
```

```python
test_df = pd.DataFrame(df["location"].value_counts())
test2_df = test_df.reset_index(level=0)
test2_df.head()
```

```python
test3_df = test2_df[test2_df.index != 13]
test3_df.head(5)
```

```python
test3_df.mean()
```

```python
locations = test3_df["index"]
locations = locations.tolist()
locations_list = [literal_eval(tup) for tup in locations]
locations_list
```

```python
lat_long_df = pd.DataFrame(locations_list, columns = ["Lat", "Long"])
lat_long_df.head()
```

```python
lat_long_df["Crash Counts"] = pd.Series(test3_df["location"])
lat_long_df.head()
```

```python
len(lat_long_df["Lat"])
```

```python
x = len(lat_long_df["Lat"])

plt.bar(x, lat_long_df["Lat"])
plt.show()
```

```python
locations = lat_long_df[["Lat","Long"]]
crashes = lat_long_df["Crash Counts"].astype(float)
```

```python
crashes.max()
```

```python
# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations[0:100], weights=crashes[0:100], dissipating=False, max_intensity= crashes.max(), point_radius=.009)

# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```

```python
# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
crash_layer = gmaps.symbol_layer(locations[0:9], fill_color = "red", stroke_color = "red", scale = 2) 
    
# Add layer
fig.add_layer(crash_layer)

# Display figure
fig
```
