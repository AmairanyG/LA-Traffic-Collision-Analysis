```python
from ast import literal_eval

locations = df["location"]
locations = locations.tolist()
locations_mod = [literal_eval(tup) for tup in locations]
locations_mod
```

```python
# Plot Heatmap
fig = gmaps.figure()

# Create heat layer
heat_layer = gmaps.heatmap_layer(locations_mod, dissipating=False, max_intensity=10, point_radius=1)

# Add layer
fig.add_layer(heat_layer)

# Display figure
fig
```
